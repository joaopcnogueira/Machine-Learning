```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
start = datetime.datetime(2019, 1, 1)
date = [start + datetime.timedelta(days=x) for x in range(90)]
```

```python
x1 = np.random.normal(0, 1, len(date))
x2 = np.random.normal(2, 4, len(date))
```

```python
df = pd.DataFrame({'Date': date, 'x1': x1, 'x2': x2})
```

```python
df.head(10)
```

```python
df.set_index('Date').plot();
```

# Usando o TimeSeriesSplit

```python
from sklearn.model_selection import TimeSeriesSplit
nsplits = 4
splits = TimeSeriesSplit(n_splits = nsplits)

plt.figure(figsize=(16,14))
layout = int(str(nsplits) + '10')
index = 1
for train_index, test_index in splits.split(df):
    train = df.loc[train_index]
    test = df.loc[test_index]
    print('Observations: %d' % (len(train) + len(test)))
    print('Training Observations: %d' % (len(train)))
    print('Testing Observations: %d' % (len(test)))
    plt.subplot(layout + index)
    plt.plot(train['Date'], train['x1'])
    plt.plot(train['Date'], train['x2'])
    plt.plot(test['Date'], test['x1'])
    plt.plot(test['Date'], test['x2'])
    index += 1
```
