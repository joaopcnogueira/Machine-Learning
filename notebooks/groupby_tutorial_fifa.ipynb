```python
fifa_19_players_data = pd.read_csv("data/fifa19.csv", )
fifa_19_players_data.drop(['Unnamed: 0', 'ID', 'Photo', 'Flag', 'Club Logo'], axis='columns', inplace=True)
pd.options.display.max_columns = 90
fifa_19_players_data.head(3)
```

# Customization of Plots using Built in Templates with user-defined color choices

```python
arsenal = team.get_group('Arsenal')
arsenal['Wage'] = arsenal['Wage'].str.replace('€', '').str.replace('K', '')
arsenal['Wage'] = arsenal['Wage'].astype('int')

# Horizontal Bar Plot
plt.style.use('ggplot')
ars = arsenal.sort_values('Wage', ascending=False).head(10)
ax = ars.plot(kind='barh', x='Name', y='Wage', color='#61D199', title='Top 10 Earners at Arsenal Football Club 2019')
ax.set_xlabel('Weekly Earnings')

# Dataframe Wage Column Bar Plot
arsenal[['Name', 'Value', 'Wage', 'Nationality']].head().style.bar(subset='Wage', color='61D199', align='zero')
```

# The .agg() Method on a groupby object

```python
team.agg({'Overall': 'mean',
          'Acceleration': 'mean',
          'ShotPower': 'mean'}).sort_values(by='Overall', ascending=False).reset_index().head(10)
```

# Create a DataFrame with the Best Player from each team

```python
team = fifa_19_players_data.groupby('Club')

greatest_club_players_df = pd.DataFrame(columns=fifa_19_players_data.columns)

for club, data in team:
    best_player = data.nlargest(1, 'Overall')
    greatest_club_players_df = greatest_club_players_df.append(best_player)
```

```python
greatest_club_players_df.sort_values('Overall', ascending=False)
```
