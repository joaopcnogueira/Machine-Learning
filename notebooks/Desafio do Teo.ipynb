```python
import pandas as pd
```

```python
dados = pd.DataFrame({"cliente":[1,1,1,2,2],
                      "mes":[1,2,3,1,2],
                      "valor":[100,250,500,350,400],
                      "parcelas":[2,5,4,1,2]})

dados
```

```python
dados = dados.assign(valor_mes = lambda x: x.valor / x.parcelas)
dados
```

```python
def enumerate_mes_parcelas(row):
    return [int(row['mes']) + i for i in range(int(row['parcelas']))]

dados = dados.assign(mes_parcelas = dados.apply(enumerate_mes_parcelas, axis=1))
dados
```

```python
dados_melt = dados.explode('mes_parcelas')
dados_melt
```

```python
final = (
    dados_melt
    .groupby(['cliente', 'mes_parcelas'])['valor_mes'].sum()
    .unstack(fill_value=0)
)
final
```
